```python
import os
import requests
import pandas as pd
import matplotlib
import json
```

```python
ALPHA_KEY = os.environ['ALPHA_KEY']
payload = {'symbol':'GGBR4.SA','function':'TIME_SERIES_DAILY_ADJUSTED', 'outputsize':'full', 'apikey':ALPHA_KEY}
alpha = requests.get('https://www.alphavantage.co/query?', params=payload)
```

```python
text_json_values = alpha.json()['Time Series (Daily)']
```

```python
df = pd.DataFrame.from_dict(text_json_values, orient='index')
```

```python
cols = ['1. open','2. high','3. low','4. close','5. adjusted close','6. volume','7. dividend amount','8. split coefficient']
df = df[cols]
```
