```python
def all_assets(Asset):
    global all_asset_codes
    all_asset_codes = []
    all_assset_info = Asset.objects.all()
    for asset in all_assset_info:     
        all_asset_codes.append(asset.asset_code)
```

```python
def call_historic_asset(asset_name):
    global time_series_asset_df
    time_series_asset_df
    ALPHA_KEY = os.environ['ALPHA_KEY']
    payload = {'symbol':asset_name,'function':'TIME_SERIES_DAILY_ADJUSTED', 'outputsize':'full', 'apikey':ALPHA_KEY}
    api_response = requests.get('https://www.alphavantage.co/query?', params=payload)
    text_json_values = api_response.json()['Time Series (Daily)']
    time_series_asset_df = pd.DataFrame.from_dict(text_json_values, orient='index')
```

```python
all_assets(Asset)
call_historic_asset('GGBR4.SA')
```
